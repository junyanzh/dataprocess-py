```python
import pandas as pd
from google.colab import drive
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())
```

```python
worksheet = gc.open('category for split').sheet1

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
#print(rows)

SP_cate= pd.DataFrame.from_records(rows)
SP_catehead = SP_cate.iloc[0]
SP_cate = SP_cate[1:]
SP_cate.columns = SP_catehead
del SP_catehead
```

```python
SP = SP_cate[['Project Uuid','Sub Category Nm']]
```

```python
catls = SP['Sub Category Nm'][2].split('、')
catls
```

```python
uuid = SP['Project Uuid'][2].split()*len(catls)
uuid
```

```python
cate_dict = {
    'Project Uuid':uuid,
    'Sub Category Nm':catls
}
cate_df = pd.DataFrame(cate_dict)
cate_df
```

```python
catls1 = SP['Sub Category Nm'][52].split('、')
uuid1 = SP['Project Uuid'][52].split()*len(catls1)
cate_dict1 = {
    'Project Uuid':uuid1,
    'Sub Category Nm':catls1
}
cate_df1 = pd.DataFrame(cate_dict1)
cate_df1
```

```python
all_df = cate_df.append(cate_df1)
all_df
```

```python
cate_df=pd.DataFrame()
for i in range(len(SP)):
  catls = SP['Sub Category Nm'].iloc[i].split('、')
  uuid = SP['Project Uuid'].iloc[i].split()*len(catls)
  cate_dict = {
    'Project Uuid':uuid,
    'Sub Category Nm':catls
    }
  temp = pd.DataFrame(cate_dict)
  cate_df = cate_df.append(temp,ignore_index=True)

```
